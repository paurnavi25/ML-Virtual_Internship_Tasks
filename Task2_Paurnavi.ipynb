```python
import pandas as pd
import io
from google.colab import files

uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))

  df = pd.read_csv(io.StringIO(uploaded[fn].decode('utf-8')))

df.head()
```

```python
data = df.copy()
data.shape
data.columns
```

```python
cols_keep = ['Restaurant Name','City','Cuisines','Average Cost for two',
             'Price range','Has Online delivery','Has Table booking',
             'Votes','Aggregate rating','Longitude','Latitude']
data = data[[c for c in cols_keep if c in data.columns]].copy()
data.head()
```

```python
df.isnull().sum()
```

```python
data['Cuisines'] = data['Cuisines'].fillna('Unknown')
data['City'] = data['City'].fillna('Unknown')
data['Average Cost for two'] = data['Average Cost for two'].fillna(data['Average Cost for two'].median())
```

```python
data['Cuisines'] = data['Cuisines'].str.lower().str.replace('/', ' ').str.replace('&', ' ')
data['Cuisines'] = data['Cuisines'].str.replace('-', ' ')
data['Cuisines'] = data['Cuisines'].str.replace('\s+', ' ', regex=True).str.strip()
```

```python
def price_bucket(x):
    if x <= 200: return 'cheap'
    if x <= 600: return 'moderate'
    return 'expensive'
```

```python
for col in ['Has Online delivery', 'Has Table booking']:
    if col in data.columns:
        data[col] = data[col].map({1:'online', 0:'offline', 'Yes':'online','No':'offline'}).fillna('unknown')
```

```python
data['price_bucket'] = data['Average Cost for two'].apply(price_bucket)
data['content'] = (data['Cuisines'].fillna('') + ' ' +
                   data['City'].fillna('') + ' ' +
                   data['price_bucket'] + ' ' +
                   data['Has Online delivery'].astype(str) + ' ' +
                   data['Has Table booking'].astype(str))

data[['Restaurant Name','content']].head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

tfidf = TfidfVectorizer(stop_words='english', ngram_range=(1,2))
tfidf_matrix = tfidf.fit_transform(data['content'])

# Cosine similarity matrix (dense can be large; for 9.5k rows it's okay but keep an eye on memory)
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
indices = pd.Series(data.index, index=data['Restaurant Name'].str.lower())
# Group by the lowercased name and take the first index for each group
indices = indices.groupby(indices.index).first()

def recommend_by_restaurant(name, top_n=10):
    name = name.lower()
    if name not in indices.index:
        print("Restaurant not found. Try a different name or check spelling.")
        return []
    idx = indices[name]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1: top_n+1]   # skip self
    restaurant_indices = [i[0] for i in sim_scores]
    return data.iloc[restaurant_indices][['Restaurant Name','City','Cuisines','Average Cost for two','Aggregate rating']]
```

```python
def recommend_by_preferences(preferred_cuisines=None, city=None, price_pref=None, top_n=10):

    pref = ''
    if preferred_cuisines:
        pref += ' '.join([c.lower() for c in preferred_cuisines]) + ' '
    if city:
        pref += city.lower() + ' '
    if price_pref:
        pref += price_pref.lower() + ' '
    # vectorize query
    q_vec = tfidf.transform([pref])
    sim_scores = cosine_similarity(q_vec, tfidf_matrix).flatten()
    top_idx = sim_scores.argsort()[::-1][:top_n]
    return data.iloc[top_idx][['Restaurant Name','City','Cuisines','Average Cost for two','Aggregate rating']]


```

```python
print("By restaurant sample:")
print(recommend_by_restaurant(data['Restaurant Name'].iloc[10], top_n=5))

print("\nBy user preferences sample:")
print(recommend_by_preferences(preferred_cuisines=['chinese'], city='mumbai', price_pref='cheap', top_n=5))
```

```python
def precision_at_k_for_restaurant(idx, k=10):
    # find top k for idx
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:k+1]
    top_idx = [i[0] for i in sim_scores]

    cuisines_set = set(data.loc[idx,'Cuisines'].split(','))
    relevant = 0
    for t in top_idx:
        if cuisines_set.intersection(set(data.loc[t,'Cuisines'].split(','))):
            relevant += 1
    return relevant / k

import numpy as np
sample_idxs = np.random.choice(data.index, 100, replace=False)
np.mean([precision_at_k_for_restaurant(i, k=10) for i in sample_idxs])
```

```python
import joblib
joblib.dump(tfidf, 'tfidf_vectorizer.pkl')
# save the data frame (lighter)
data.to_csv('recommender_data.csv', index=False)
```
